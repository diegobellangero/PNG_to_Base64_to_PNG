# Imports

```python
import base64                       # Importa biblioteca de codificação/decodificação do Base64
from PIL import Image               # Importa o módulo Image (manipulação de imagens) da biblioteca PIL (Python Image Library). O Pillow é um Fork do PIL
from io  import BytesIO             # Importa o módulo BytesIO (manipular bytes como se fossem arquivos) da bilioteca IO
from IPython.display import display # Importa o módulo display (exibir mídias) da bilioteca IPython (ambiente integrado para o Jupyter)
from IPython.display import HTML    # Importa o módulo HTML (manipulação e exibição de HTML) da bilioteca IPython (ambiente integrado para o Jupyter)
```

# Parâmetros

```python
# Base64 para PNG

# string Base64 com cabeçalho 'data:image/png;base64,'
string_base64 = ""
arquivo_saida = "imagem.png"           # Relativo ao projeto

# PNG para Base64
caminho_da_imagem = "Rick.png"         # Relativo ao projeto
nova_largura = 192                     # Largura da imagem desejada. A altura será calculada proporcionalmente
```

# Base64 para PNG

```python
def base64_to_png(base64_string, output_file):
    try:
        # Remove o prefixo, se presente
        if base64_string.startswith("data:image/png;base64,"):
            base64_string = base64_string[len("data:image/png;base64,"):]

        # Decodifica a string Base64 em bytes
        image_data = base64.b64decode(base64_string)

        # Cria uma instância da classe Image usando o BytesIO
        img = Image.open(BytesIO(image_data))

        # Salva a imagem como arquivo PNG
        img.save(output_file, "PNG")
        
        print(f"Imagem convertida e salva como {output_file}")
        display(Image.open(output_file))
    except Exception as e:
        print("Ocorreu um erro:", e)

base64_to_png(string_base64, arquivo_saida)
```

# PNG para Base64

```python
def resize_image(image_path, new_width):
    try:
        img = Image.open(image_path)
        aspect_ratio = img.width / img.height
        new_height = int(new_width / aspect_ratio)
        
        resized_img = img.resize((new_width, new_height), Image.LANCZOS)
        # Ou: resized_img = img.resize((new_width, new_height), Image.resampling.lanczos)
        
        return resized_img
    except Exception as e:
        print("Ocorreu um erro:", e)
        return None

def png_to_base64(image_path, new_width):
    try:
        resized_img = resize_image(image_path, new_width)
        
        if resized_img:
            buffered = BytesIO()
            resized_img.save(buffered, format="PNG")
            base64_data = base64.b64encode(buffered.getvalue()).decode("utf-8")
            
            return f"data:image/png;base64,{base64_data}"
    except Exception as e:
        print("Ocorreu um erro:", e)
        return None

# Substitua 'caminho_da_imagem.png' pelo caminho do seu arquivo PNG

base64_string = png_to_base64(caminho_da_imagem, nova_largura)

if base64_string:
    display(HTML(f'<img src="{base64_string}">'))
    print("String Base64 da imagem redimensionada com prefixo:")
    print(base64_string)
```
